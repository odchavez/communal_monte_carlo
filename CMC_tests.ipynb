```python
import numpy as np

import particle_filter
import simulate_data

M=1
params={'N': 50, 'N_batch': 5000, 'omega_shift' : [0,3], 'shards': M}
PART_NUM=1000
test = simulate_data.simulated_data(params, model="probit_sin_wave").get_data()
#print('test keys:', test.keys())
#test[list(test.keys())[0]]
```

```python
pf_obj=list()
for m in range(M):
    pfo = particle_filter.particle_filter(test['shard_'+str(m)], PART_NUM, model='probit_sin_wave', sample_method="importance")
    pf_obj.append(pfo)
    pf_obj[m].run_particle_filter()
```

```python
particle_0_7=pf_obj[0].get_particle(7)
particle_0_7.plot_parameters()
```
